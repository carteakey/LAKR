```python
import datasets
from datasets import load_dataset
# Use cudf
%load_ext cudf.pandas
```

```python
dataset = load_dataset("McAuley-Lab/Amazon-Reviews-2023", "raw_review_Video_Games", trust_remote_code=True)
```

```python
dataset_2 = load_dataset("McAuley-Lab/Amazon-Reviews-2023", "raw_meta_Video_Games", trust_remote_code=True)
```

```python
# Coverting the dataset to pandas dataframe
import pandas as pd

df = pd.DataFrame(dataset['full'])
df_2 = pd.DataFrame(dataset_2['full'])
```

```python
# Save 1000 rows of the dataset to a csv file

df[:1000].to_csv('amazon_reviews.csv', index=False)
df_2[:1000].to_csv('amazon_reviews_meta.csv', index=False)
```

```python
reviews_df = df
meta_df = df_2 

# Convert timestamp to datetime
reviews_df['timestamp'] = pd.to_datetime(reviews_df['timestamp'], unit='ms')

# Summary statistics for numeric columns in meta_df
meta_numeric_summary = meta_df.describe()

# Summary statistics for numeric columns in reviews_df
reviews_numeric_summary = reviews_df.describe()

# Summary statistics for categorical columns in meta_df
meta_categorical_summary = meta_df.describe(include=['O'])

# Summary statistics for categorical columns in reviews_df
reviews_categorical_summary = reviews_df.describe(include=['O'])

# Checking the distribution of null values in meta_df
meta_null_distribution = meta_df.isnull().sum()

# Checking the distribution of null values in reviews_df
reviews_null_distribution = reviews_df.isnull().sum()

(meta_numeric_summary, meta_categorical_summary, meta_null_distribution,
 reviews_numeric_summary, reviews_categorical_summary, reviews_null_distribution)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Remove columns with entirely missing values
meta_df_cleaned = meta_df.drop(columns=['bought_together', 'subtitle', 'author'])

# Impute missing values for 'main_category' and 'store' with the mode
meta_df_cleaned['main_category'].fillna(meta_df_cleaned['main_category'].mode()[0], inplace=True)
meta_df_cleaned['store'].fillna(meta_df_cleaned['store'].mode()[0], inplace=True)

# Function to plot distributions of numeric variables
def plot_distribution(df, column, title, xlabel, ylabel):
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], kde=True)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

# Plot distributions for meta_df
plot_distribution(meta_df_cleaned, 'average_rating', 'Distribution of Average Rating', 'Average Rating', 'Frequency')
plot_distribution(meta_df_cleaned, 'rating_number', 'Distribution of Rating Number', 'Rating Number', 'Frequency')

# Plot distributions for reviews_df
plot_distribution(reviews_df, 'rating', 'Distribution of Ratings', 'Rating', 'Frequency')
plot_distribution(reviews_df, 'helpful_vote', 'Distribution of Helpful Votes', 'Helpful Votes', 'Frequency')
```

```python
# Function to plot bar charts for categorical variables
def plot_bar_chart(df, column, title, xlabel, ylabel):
    plt.figure(figsize=(12, 8))
    sns.countplot(y=df[column], order=df[column].value_counts().index)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

# Plot bar charts for key categorical variables in meta_df
plot_bar_chart(meta_df_cleaned, 'main_category', 'Frequency of Main Categories', 'Frequency', 'Main Category')
plot_bar_chart(meta_df_cleaned, 'store', 'Frequency of Stores', 'Frequency', 'Store')
```

```python
# Convert timestamp to datetime
reviews_df['timestamp'] = pd.to_datetime(reviews_df['timestamp'], unit='ms')

# Plot the number of reviews over time
plt.figure(figsize=(14, 8))
reviews_df.set_index('timestamp').resample('M').size().plot()
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
# Merge reviews_df with meta_df_cleaned on 'parent_asin'
merged_df = pd.merge(reviews_df, meta_df_cleaned, left_on='parent_asin', right_on='parent_asin')

# Display the first few rows of the merged dataframe
print(merged_df.head())
```

```python
# Compute the correlation matrix
correlation_matrix = merged_df[['average_rating', 'rating_number', 'rating', 'helpful_vote']].corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```
